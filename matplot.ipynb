```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.figure(figsize = (12,8))
plt.show()
```

```python
x = [0,2,4,6,8]
y=[0,4,8,12,16]
plt.plot(x,y)
```

```python
x = np.arange(0,6)
y=4*x-2
plt.plot(x,y , "--")
```

```python
x = np.arange(0,20)
y = 2*x+3
y2 = 4*y-3
y3 = x/2+8
plt.plot(x,y , "--", label="2x+3")
plt.plot(y,y2 , label ="4y-3")
plt.plot(x,y3 , label="x/2+8")
plt.legend()
```

```python
x = np.arange(0,100) /10
y = np.sin(x)
z = np.cos(x)
t = np.sin(y)
v = np.cos(y)
plt.plot(x,y,color="red",label ="sin(x)")
plt.plot(x,z ,color="blue",label ="cos(x)")
plt.plot(x,t,color="purple",label ="sin(y)",linestyle ="-")
plt.plot(x,v ,color="green",label ="cos(y)")



plt.legend()
plt.show()
```

```python
data = {
    "Yıl": np.arange(2010, 2021),   # 2010–2020 (11 yıl)
    "Satış": np.arange(400, 520, 11),  # 11 değer (100 → 210)
    "Masraf": np.random.randint(50, 400, 11)  # 11 değer
}

df = pd.DataFrame(data)
regions = ["Kuzey","Güney","Doğu","Batı"]
df["Bölge"] = np.random.choice(regions, df.shape[0])

plt.plot(df["Yıl"], df["Masraf"],
         label="Satışlar",
         marker="o")

plt.legend()
```
